```python
#1

import random
d1 = []
for i in range(1000):
    d1.append(random.randint(60, 70))
print(d1)
```

```python
import pandas as pd

df = pd.DataFrame(d1)
df
```

```python
# 2
mean = df.mean()
print("Mean value: ",mean)
```

```python
std = df.std()
print("Standard-deviation: ",std)
```

```python
# 3
import matplotlib.pyplot as plt

plt.hist(d1 , bins=10, color='green', edgecolor='black', linewidth=1.2)
plt.title('Histogram of 1000 Random Values between 60 and 70')
plt.show()
```

```python
#4 and 5
import numpy as np
sample=[]
Smean=[]
for j in range(30):
    for i in range(30):
        j=random.choice(d1)
        sample.append(j)
    s=np.mean(sample)
    Smean.append(s)
print(Smean)
```

```python
#6
import numpy as np
sample=[]
Smean=[]
for j in range(400):
    for i in range(50):
        j=random.choice(d1)
        sample.append(j)
    s=np.mean(sample)
    Smean.append(s)
print(Smean)
```

```python
#7
plt.hist(Smean , bins=10, color='green', edgecolor='black', linewidth=1.2)

plt.show()
```

```python
#8
print("Sample mean: ",np.mean(Smean))
```

```python
print('Standard deviation: ', np.std(Smean))
```
